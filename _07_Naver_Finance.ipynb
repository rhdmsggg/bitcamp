```python
import urllib.request as req
from bs4 import BeautifulSoup
```

```python
url = "https://finance.naver.com/"
```

```python
# 1. 요청 후 객체를 전달
html = req.urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
html
```

```python
# 2. 요청후 문자열로 전달
html = req.urlopen(url).read()
soup = BeautifulSoup(html,'html.parser')
html
```

```python
usd_price = soup.select_one("#content > div.article2 > div.section1 > div.group1 > table > tbody > tr.down.bold > td:nth-of-type(1)")
usd_price
```
